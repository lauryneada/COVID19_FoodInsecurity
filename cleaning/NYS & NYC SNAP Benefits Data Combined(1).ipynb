```python
snap_population = pd.read_csv("Borough_Community_District_Report_-_SNAP_Population-6-1.csv")
snap_population
```

```python
snap_population['First Day of the Report Month']= pd.to_datetime(snap_population['First Day of the Report Month'])
snap_population
```

```python
snap_population['First Day of the Report Month'] = snap_population['First Day of the Report Month'].apply(str)
```

```python
snap_population_state = pd.read_csv("Supplemental_Nutrition_Assistance_Program__SNAP__Caseloads_and_Expenditures__Beginning_2002-4.csv")
snap_population_state
```

```python
snap_population_state.drop(['Unnamed: 0'], axis = 1, errors='ignore')
```

```python
snap_population.rename(columns = {'Individuals Receiving SNAP Benefits':'Total SNAP Persons'}, inplace = True)
snap_population
```

```python
snap_population.rename(columns = {'Borough':'District'}, inplace = True)
snap_population
```

```python
snap_population.rename(columns = {'First Day of the Report Month':'Year'}, inplace = True)
snap_population
```

```python
snap_population = snap_population.replace("QUEENS", "Queens")
snap_population
```

```python
snap_population = snap_population.replace("MANHATTAN", "New York")
```

```python
snap_population = snap_population.replace("BROOKLYN", "Kings")
```

```python
snap_population = snap_population.replace("STATEN ISLAND", "Richmond")
```

```python
snap_population = snap_population.replace("BRONX", "Bronx")
```

```python
snap_population
```

```python
snap_population.drop(['Unnamed: 0'], axis = 1, errors='ignore')
```

```python
df = pd.concat([snap_population_state, snap_population], ignore_index=True)
```

```python
df = df.drop(['Unnamed: 0'], axis = 1, errors='ignore')
```
