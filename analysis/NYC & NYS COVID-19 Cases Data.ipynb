```python
covid_case = pd.read_csv("COVID-19_Daily_Counts_of_Cases__Hospitalizations__and_Deaths.csv")
covid_case
```

```python
covid_case.rename(columns = {'QN_CASE_COUNT':'Number of Cases in Queens'}, inplace = True)
covid_case
```

```python
covid_case.rename(columns = {'BK_CASE_COUNT':'Number of Cases in Brooklyn'}, inplace = True)
covid_case
```

```python
covid_case['Date']= pd.to_datetime(covid_case['Date'])
covid_case
```
